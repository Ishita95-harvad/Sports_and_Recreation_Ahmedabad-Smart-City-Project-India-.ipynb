## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and removing duplicates.


**Reasoning**:
Clean the data by handling inconsistencies in the 'Phone No. / Timings' and 'Area' columns, and remove duplicate rows.



```python
import re

# Clean 'Phone No. / Timings' column
def clean_phone_number(text):
    if pd.isna(text):
        return text
    match = re.search(r'(\d{10,})', str(text))
    if match:
        return match.group(1)
    else:
        return text

df['Phone No. / Timings'] = df['Phone No. / Timings'].apply(clean_phone_number)

# Clean 'Area' column
df['Area'] = df['Area'].str.title()

# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Report the cleaning process
print("Cleaning Steps:")
print("1. Phone Number Cleaning: Extracted phone numbers using regular expressions, and preserved existing timings. Any non-numerical value is preserved.")
print("2. Area Cleaning: Converted area names to title case.")
print("3. Duplicate Removal: Removed any rows that are identical in all columns.")

print(f"\nShape of the cleaned DataFrame: {df_cleaned.shape}")
display(df_cleaned)
```

## Data wrangling

### Subtask:
Prepare the data for analysis and visualization.


**Reasoning**:
Convert the 'Area' column to categorical, clean the 'Phone No. / Timings' column, and create the 'FacilityType' column based on keywords in the 'Name of the Sports Complex' column.



```python
import pandas as pd

# Convert 'Area' to categorical
df_cleaned['Area'] = df_cleaned['Area'].astype('category')

# Further clean 'Phone No. / Timings'
def clean_phone_number_further(text):
    if pd.isna(text) or isinstance(text, int):  # Handle potential int type
        return text
    text = str(text).split('/')[0].split(',')[0].strip()  # Extract the first part of the phone number
    text = re.sub(r'[^\d\s]', '', text)  # Remove non-numeric characters except space
    return text

df_cleaned['Phone No. / Timings'] = df_cleaned['Phone No. / Timings'].apply(clean_phone_number_further)

# Create 'FacilityType' column
keywords = ["Stadium", "Ground", "Pool", "Club", "Arena"]
def extract_facility_type(name):
    name = str(name).lower()  # Convert name to lowercase
    for keyword in keywords:
        if keyword.lower() in name:
            return keyword
    return "Other"

df_cleaned['FacilityType'] = df_cleaned['Name of the Sports Complex'].apply(extract_facility_type)

display(df_cleaned.head())
```

## Data clustering

### Subtask:
Perform K-means clustering on the data to group similar sports and recreation facilities.


**Reasoning**:
Perform K-means clustering on the data using selected features, scale the features, determine the optimal number of clusters, and visualize the clusters.



```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Select features for clustering
features = ['Area', 'FacilityType']
X = pd.get_dummies(df_cleaned[features], drop_first=True)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using the elbow method (adjusting the range)
wcss = []
for i in range(1, min(4, len(X_scaled))):  # Limit the range to avoid errors
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, min(4, len(X_scaled))), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Apply KMeans with k=2 (adjusting k)
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
df_cleaned['cluster'] = kmeans.fit_predict(X_scaled)

# Visualize clusters using PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
df_cleaned['PC1'] = principal_components[:, 0]
df_cleaned['PC2'] = principal_components[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=df_cleaned, palette='viridis')
plt.title('KMeans Clustering Visualization with PCA')
plt.show()
```
